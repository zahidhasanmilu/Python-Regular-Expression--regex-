```python
import requests
from bs4 import BeautifulSoup

# Local file path
file_path = r"C:\Users\Zahid Hasan Milu\Desktop\NEW PROJECT\Python\test.html"

# Open and read the local HTML file
with open(file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

print("#------------------------------")
# Assuming there is at least one <a> tag in the body
if soup.body.a:
    for i in soup.body.find_all('a'):
        # Prints the href attribute of the first <a> tag in the body
        print(i['href'])
        print(i.text)       # Prints the text of the first <a> tag in the body
        # Prints the target attribute of the first <a> tag in the body, or 'No target' if not present
        print(i.get('target', 'No target'))
else:
    print("No <a> tag found in the body")

print("#------------------------------")
# Assuming there is at least one <ol> tag in the body
if soup.body.ol:
    # Prints the text of the first <ol> tag in the body, stripping any leading/trailing whitespace
    print(soup.body.ol.text.strip())
else:
    print("No <ol> tag found in the body")

print("#------------------------------")
# Assuming there is at least one <ul> tag in the body
if soup.body.ul:
    # Prints the text of the first <ul> tag in the body, stripping any leading/trailing whitespace
    print(soup.body.ul.text.strip())
else:
    print("No <ul> tag found in the body")
```

```python
import bs4

file_url = r"C:\Users\Zahid Hasan Milu\Desktop\NEW PROJECT\Python\test.html"

# Read the content of the HTML file
with open(file_url, 'r', encoding='utf-8') as file:
    content = file.read()

# Parse the HTML content using BeautifulSoup
soup = bs4.BeautifulSoup(content, 'html.parser')

# Find all <a> tags
content_list = soup.find_all('a')

# Initialize the counter outside the loop
total = 0

# Loop through all <a> tags and print the details
for i in content_list:
    total += 1
    print(
        f'Number of tag: {total}, Tag: {i.name}, Text: {i.text}, Target attribute: {i.get("target")}')

# Print the total number of <a> tags
print(f'Total number of <a> tags: {total}')
```
